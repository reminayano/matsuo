## Section3 テクニック・発展内容
### 3.1 前処理
学習の高速化や性能向上等の為に入力データを変換する前処理を行うことがある。

前処理の1つに**スケーリング**がある。スケーリングとは、特定の範囲にデータが収まるように変換すること。

例えば、mnistの入力画像の各ピクセルの値は0～255の範囲の整数値だが、それぞれ255で割ることにより、値を0～1の浮動小数点に収めることが出来る。

```python
x_train, x_test = x_train/255, x_test/255
```

その他の前処理には、**正規化**(平均や分散を特定の値にする)、**白色化**(データの各要素間の相関を取り除く)、**バッチ正規化**(データ全体を考慮)等がある。

### 3.2 勾配に関するテクニック
MLPにおける最大の目的はモデルのパラメータを適切に学習すること。

つまり、パラメータの更新量が適切に定まることが最も重要であるといえる。

実は、このパラメータの更新量は**勾配**(＝損失関数のパラメータに関する微分)によって決定される。

この勾配は過大・過小になりやすいので、適切に扱う必要がある。

ここでは、パラメータ更新量や勾配を適切にするためのテクニックを3つ(optimizer, activation, initialization)見ていく。

#### 3.2.1 最適化アルゴリズム(optimizer)

まずは、パラメータの学習に直接効いてくる、パラメータの更新方法、つまり、最適化アルゴリズムから見ていく。

最適化アルゴリズムには様々なものがあり、kerasでも多くのoptimizerを選べる。

最も基本的な最適化アルゴリズムは**最急降下法**と、 **確率的勾配降下法**(**Stochastic Gradient Descent, SGD**)である。

**最急降下法**
再急降下法(勾配降下法ともいう)は、１階の微分でアル勾配のみを用いて最適化を行う。

パラメータ$w$の$t$回目の更新を、損失関数$E$として、

$$
w^{(t+1)} \leftarrow w^{(t)} - \eta \nabla E(w^{(t)}) (\nabla E(w^{(t)} = \frac{\partial E(w)}{\partial w} |_{w=w_{(t)}} )
$$


∇(ナブラ) ：ベクトル微分演算を表す。
∇=(∂/∂x, ∂/∂y, ∂/∂z)的な。

のように行う。

$\eta$は**学習率**ととよばれるパラメータで、明示的に指定する必要がある。

なお、最急降下法では損失関数の計算には常に全訓練データを用い、

$$
E(W^{(t)})=\Sigma_n E_n(w^{(t)})
$$

として計算する。($E_n$はn番目の訓練データによる損失関数)

上記の更新式のイメージ
- $\nabla E(w^{(t)})$ : パラメータの空間で損失関数が最も増加する方向(＝**勾配**)
- $\eta$ : パラメータの空間で損失関数が最も
減少する方向 $-\nabla E(w^{(t)})$ にどれだけ進のかという比率

最急降下法はほとんどの最適化手法の基本形になっているので、イメージを抑えておくこと。

**SGD**

SGDは最も一般的なMLPのオンライン学習手法。

パラメータの更新を
$$
w^{(t+1)} \leftarrow w^{(t)} - \eta \nabla E(w^{(t)}) ( \nabla E(w^{(t)})
 = \frac{\partial E^{(t)}(w)}{\partial w} |_{w=w_{(t)}} )
$$

のように行う。

ここで、
$$
E(w^{(t)})= \frac {1}{|B^{(t)}|}\Sigma_{n\in B^{(t)}}E_n(w^{(t)})
$$


である。

このように、各更新毎にランダムにミニバッチ(＝訓練データの抜粋、部分集合) $B^{(t)}$を構成し、その平均損失を使用することがSGDの特徴。

Kerasでは以下のクラスを使用する。

```python
keras.optimizers.SGD(lr=0.01, momentum=0.0, decay=0.0, nesterov=False)
```

主な引数
- lr : 学習率、0以上の実数
- momentum : モーメンタム、0以上の実数(前回のパラメータ更新量を反映させる比率)
- decay : 更新毎の学習率の減衰率、0以上の実数
- nesterov : Nesterov momentumを適用するかどうか(True→モーメンタム項の計算を1ステップ先読みして評価する)

keras.optimizers.SGD()はcompile関数の引数でoptimizer='sgd'と指定することで使用できる。

もしデフォルトでない引数でSGDを使用する場合する場合はoptimizer=keras.optimizer.SGD(lr=0.01, momentum=0.9, nestrov=True)などとすればよい。

**その他の最適化アルゴリズム**

SGDでは学習率は明示的に与える必要があった。しかし。どのような学習率が最適なのか？

学習率が大きければ、勾配方向への移動距離が大きくなり、より早く学習できそう。

しかし、学習が進むと、次第に「最適解」にちkづくことが期待されるが、このとき学習率が大きすぎると「通り過ぎる」可能性もある。

従って、学習の序盤は大きな学習率で早く解の付近にたどり着き、次第に学習率を減らして、慎重に解に位置を合わせていきたい。

しかし、このように学習率の調整を手動で行うのは困難→学習率の調整が組み込まれた最適化手法が複数提案されている。

また、単純なSGDでは勾配が小さい平坦な場所に来ると更新がほとんど行われなくなってしまうという短所もあるが、この点についても同時に対策が成されている。

kerasでは後述するような、より便利な最適化アルゴリズムも、compile時にoptimizerパラメータを変更するだけで簡単に使用可能である。

(4つほど最適化手法を紹介しているがとりあえず飛ばすこととする。時間がある時に。)

#### 3.2.2 活性化関数(activation)

勾配に関するテクニックには、上記のほかに、活性化関数の選択がある。

前述の最適化手法で見た通り、勾配がパラメータ更新で中心的な役割を果たし、その大きさが適切に保たれることは極めて重要である。

しかし、一方でこの勾配の大きさについ**て勾配消失問題**という課題があり、活性化関数の工夫はその対処法の1つとなっている。

勾配消失問題はなぜ発生するのか？⇒ニューラルネットワークの勾配の求め方である**誤差逆伝播法**を知ることで理解できる。

簡単に言うと、誤差逆伝播法では、**(ある層の勾配)＝(1層前の勾配)×(2層前の勾配)×・・・(出力層の勾配)** と積の形で勾配を求めるため、途中の勾配が小さいと入力層付近の勾配はどんどん0に近づいて行ってしまう。

従って、勾配消失問題は特に多層な場合に致命的となる。

![backprop](https://github.com/reminayano/matsuo/blob/master/lesson1/figures/backprop.png?raw=true)

ただし、勾配を求めるにあたって、差分による数値微分などの他の手法を利用することは計算量的に現実的でない。

そこで、勾配消失問題への対処の1つとして、活性化関数の微分の値がなるべく小さくならないようなものを選びたい。

例えば、全Sectionで活性化関数として扱った中で、sigmoid、tanh、ReLU関数の微分はそれぞれ以下の様になる。


````python
import numpy as np

def deriv_sigmoid(x):
    return np.exp(x)/(1+np.exp(x))**2

def deriv_tanh(x):
    return 1 - np.tanh(x)**2

def deriv_relu(x):
    return 1 * (x > 0)
````

これらの3種の微分をグラフで表示することで、勾配(微分)がどのように変化するか見てみる。

````python
%matplotlib inline
import matplotlib.pyplot as plt

fig = plt.figure()
x = np.linspace(-10, 10, 1000)

# sigmoid
ax = fig.add_subplot(111)
ax.plot(x, deriv_sigmoid(x), label='sigmoid deriv')

# tanh
ax.plot(x, deriv_tanh(x), label='tanh deriv')

# relu
ax.plot(x, deriv_relu(x), label='ReLU deriv')

plt.legend()
plt.xlim(-2, 2)
plt.ylim(0, 1.2)
plt.show()
````

上図より、sigmoidよりもtanh、reluのほうがより大きな値を取り、勾配消失しにくいことが分かる、

論文でも、reluもしくはその派生形を用いているものが多い。

#### 3.2.3 初期化(initializer)

勾配に関するテクニックの3つ目はパラメータの初期化について。

各層のパラメータは0を中心とした乱数で初期化するが、大きすぎる値で初期化すれば学習の初期段階での勾配が過大になり、逆に小さすぎる値だと、勾配自体も過小になってしまい、いずれにしても学習はうまく進まない。

パラメータの初期化について、比較的頻繁に用いられる方法として、LeCunによる手法、Glorotによる手法、Heによる手法があげられる。

特に、Glorotの初期化法は活性化関数が全て線形な場合の解析結果であり、中央付近が線形とみなせるsigmoid関数やtanh関数に適しているといえる。

また、Heの初期化法は活性化関数がReLUの時に適している。

### 3.3 過学習に対するテクニック

ここで、MLPの教師あり学習における真の目標を振り返っておくと、新しく未知のデータが来た時に正確に予測することであった。

一方で、これまで扱ってきたのは、学習＝既知のデータに対して正確に予測すること。

この違いは実際大きな問題で、既知のデータへの
予測精度を十分高めたからと言って、未知のデータへの予測精度が上がるとは限らない。

具体的には、学習がある程度進むと、次第に既知データが持つ、それ自身には意味がないような統計的なばらつきまで学習してしまう。

**過学習**が起こる。すなわち、未知データへの予測精度が落ちるという現象が起きてしまう。

従って、単に目的関数を訓練データに対して最小化すればいいという問題ではなくなってくる。

こうした過学習に対するアプローチ方法として例えば以下の3つがあげられる。

#### 3.3.1 正則化(regulation)

過学習が発生する1つの理由は、MLPのモデルは特に多くのパラメータ・自由度を持つために、

訓練データに対して、その本質的な部分以上に統計的ばらつきまで含めて完全にフィットしようとしてしまうことにある。

そこで、過学習を回避するには、学習過程でいくつかのパラメータが自動的に機能しなくなると良いのだが、これを実現するのが**正規化**である。

具体的には、損失関数にパラメータの大きさに対するペナルティ項(正規化項)を含める。

これにより、パラメータを自由に動けないように制限し、なるべく少ないパラメータでデータにフィットするようにできる。

(実際の挙動としては、パラメータの値が0に近づくようにするということ。0であればパラメータは実質的に機能しない)

なお、正規化には様々な種類が存在あし、主にL2/L1正規化、またはそれらを組み合わたElasticNetが用いられる。

(L2正規化、L1正規化、ElasticNetの説明、略)

#### 3.3.2 早期終了 (early stopping)

下図のように、過学習はある一定以上学習が進んでから発生する。

また、このとき、学習データによる評価関数値が改善しても、検証データによる評価関数値はさほど改善しないか悪化する。

<img src=figures/earlystopping.png>

つまり、過学習に突入してからの学習は、未知のデータに対する予測力に寄与しない、(統計的なばらつき等の)本質的でない部分の学習である。

そこで、いっそのこと早々に学習を止めてしまうことで過学習を回避する手もある。(＝**早期終了(early stopping)**)

具体的にいつ学習を止めるかが重要である、

前述したように、訓練データに対する誤差は最適化によりほとんどの場合低下するが
、検証データの誤差は過学習時に上昇する。

そこで、検証データの誤差が大きくなってきた(あるいは評価関数値が下がってきた)ところで学習をストップさせる。

kerasではcallbackという機能をつかって各エポック毎のモデルにパラメータを保存し、検証データのコストが大きくなったら前のエポックのパラメータを使用するようにする。

model.fitの引数に以下のように設定する。

```python
model.fit(x=x_train, y=y_train, ...,
    callbacks=keras.callbacks.EarlyStopping(patience=0, verbose=1))
```

#### 3.3.3 ドロップアウト(dropout)

過学習で余計に学習している部分に着目すると、それは統計的なばらつきといえるような部分であった。

こうした確率的なばらつきは、一般に大量に足し合わせることで相殺できる。

そこで、訓練データセットから部分訓練データセットを大量に作成し、各モデルの予測結果を平均するアンサンブルという手法が用いられることがある。

このアンサンブル法は、大変魅力的な手法であるが、とてつもない計算量を要するのでそのまま用いるのは難しい。

そこで、**ドロップアウト(dropout)**とよばれる手法を用いる。ドロップアウトは近似的にアンサンブル法を実現する。

具体的には、ドロップアウトは入力の一部をランダムに0にして出力するlayerの一種。要するに、一部のユニットを取り除いた状況を再現する。

このユニットの除去を確率的に行い、一部のユニットが除去された部分ネットワークに対して学習することを繰り返すことで、多数のモデルを同時に訓練することと同じ効果を実現している。

kerasではkeras.layer.core.Dropoutクラスを用いて実装できる。

主な引数
- rate : 入力を0にする確率、0～1の実数
- seed : 乱数のシード値

(入力と出力でshapeは変わらない)

```python
keras.layers.core.Dropout(rate, noise_shape=None, seed=None)
```
